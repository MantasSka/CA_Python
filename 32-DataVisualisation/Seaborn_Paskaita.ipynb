```python
%matplotlib inline
```

```python
tips = pd.read_csv("tips.csv")
```

```python
sns.histplot(tips['total_bill'])
```

```python
sns.histplot(tips['total_bill'], kde=True)
```

```python
sns.histplot(tips['total_bill'], kde=True, binwidth=30)
```

```python
sns.histplot(data=tips, x='total_bill', kde=True, binwidth=30, hue="sex")
```

```python
g = sns.histplot(data=tips, x='total_bill',kde=True, hue="sex", multiple="stack")
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, hue="smoker")
```

```python
sns.scatterplot(x='tip', y='total_bill', data=tips, hue="sex", size="total_bill", palette='cividis')
```

```python
correlate = tips.corr()
```

```python
correlate
```

```python
sns.heatmap(correlate, annot=True)
```

```python
g = sns.FacetGrid(data = tips, col='time', row='sex', hue='sex', sharey=False) 
g.map(sns.scatterplot, 'tip', 'total_bill')

g.set_axis_labels('Tips given', "Total bill")
g.set_titles(template='{row_name} {col_name} informacija' )
```

```python
sns.set_style('white')
sns.barplot(x='sex', y='total_bill', data=tips, hue='day', estimator=sum, palette='YlOrBr')
sns.despine()
```
