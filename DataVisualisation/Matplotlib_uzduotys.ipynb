## 3 Atspausdinkite kelias linijas kurios parodo horsepower, weight ir acceleration. Atspausdinkite pirmus 10. Formatavimą galite daryti pagal save

```python
import pandas as pd

data = pd.read_csv('mpg.csv').head(10)
horsepower = data['horsepower']
weight = data['weight'] 
acceleration = data['acceleration']

plt.plot(horsepower, label="Horsepower", color="r", linestyle="--", marker=".")
plt.plot(weight, label="Weight", color="b", linewidth=10)
plt.plot(acceleration, label="Tip", color="g")

plt.legend()
```

## 4 Atspausdinkite kelis stulpelinius grafikus kurie parodo horsepower, weight ir displacement. Atspausdinkite pirmus 10. Formatavimą galite daryti savaip

```python
import pandas as pd

data = pd.read_csv('mpg.csv').head(10)

horsepower = data['horsepower']
weight = data['weight'] 
displacement = data['displacement']
model_year = data['model_year']

width=0.1

plt.bar(horsepower.index-width*2, horsepower, label="Horsepower", color="r", width=width)
plt.bar(weight.index-width, weight, label="Weight", color="b", width=width)
plt.bar(displacement.index+width, displacement, label="Displacement", color="g", width=width)
plt.bar(model_year.index+width*2, model_year, label="Model year", color="y", width=width)

plt.legend()
```

## 5 Atspausdinkite skritulinę diagrama kuri procentais parodo kiek mašinų pagaminta USA ir kiek Japonijoje. Patarimas - galite naudoti pandas filtrą ir count

```python
data = pd.read_csv('mpg.csv')

usa_filter = (data['origin'] == 'usa')
japan_filter = (data['origin'] == 'japan')

usa = data.loc[usa_filter, 'origin']
japan = data.loc[japan_filter, 'origin']

values = [usa.count(), japan.count()]
labels = ["USA", "JAPAN"]
explode = [0,0.1]

plt.pie(values, startangle=90, labels=labels, explode=explode, autopct='%1.1f%%');
```

## 6 Atspausdinkite skritulinę diagrama kuri procentais parodo model_year pasiskirstymą

```python
data = pd.read_csv('mpg.csv')

# values = data.groupby(['model_year']).sum()

values = data['model_year'].value_counts().tolist()
labels = data['model_year'].value_counts().index.tolist()

print(values)
print(labels)

plt.pie(values, labels=labels,  autopct='%1.1f%%');
```

## 7 Atspausdinkite skritulinę diagrama kuri skaičiais parodo country pasiskirstymą

```python
data = pd.read_csv('mpg.csv').head(20)

values = data['origin'].value_counts().tolist()
labels = data['origin'].value_counts().index.tolist()

print(values)
print(labels)

total = sum(values)
plt.pie(values, labels=labels, autopct=lambda p: '{:.0f}'.format(p * total / 100))
```

## 8 Atspausdinkite histogramas kuri parodo aksleracijos pasiskirstymą mašinų su trimis cilindrais

```python
data = pd.read_csv('mpg.csv')

grouped = data.groupby('cylinders')
four_cylinders_acc = grouped['acceleration'].get_group(3)

plt.hist(four_cylinders_acc, label="Akcleracija - 3 cilindrai", edgecolor="black", facecolor='g')

plt.legend()
plt.grid(True)
```

## 9 Atspausdinkite sklaidos diagramą kurioje parodomas saryšis tarp displacement ir aksceleracijos

```python
data = pd.read_csv('mpg.csv')

origin = data['origin'] 
acceleration = data['acceleration'] 
displacement = data['displacement'] 

plt.scatter(x=displacement, y=acceleration, s=displacement, c=acceleration, cmap="Reds" )
plt.title("Scatter plot")
plt.xlabel("Displacement")
plt.ylabel("Acceleration")
```
