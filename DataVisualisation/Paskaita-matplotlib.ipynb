```python
import pandas as pd

data = pd.read_csv('tips.csv')
total_bill = data['total_bill'].head(20) # paimame pirmus 10 duomenų
tip = data['tip'].head(20) # paimame pirmus 10 duomenų

width=0.5

plt.bar(total_bill.index, total_bill, label="Total Bill", color="r", width=1, edgecolor="black")
plt.bar(tip.index, tip, label="Tip", color="b", width=width, edgecolor="black")
plt.legend()
```

```python
import pandas as pd

data = pd.read_csv('tips.csv')
total_bill = data['total_bill'].head(20)
tip = data['tip'].head(20)

width=0.25 # sutvarkome plotą

plt.bar(total_bill.index-width, total_bill, width=width, label="Total Bill", color="r")
plt.bar(tip.index+width, tip, width=width, label="Tip", color="b")
plt.legend()
```

```python
data = pd.read_csv('tips.csv')
total_bill = data['total_bill']
tip = data['tip']
values = [sum(total_bill), sum(tip)]
labels = ["Total bill", "Tips"]
explode = [0,0.1]

plt.pie(values, startangle=90, labels=labels, explode=explode, autopct='%1.1f%%')
```

```python
data = pd.read_csv('tips.csv')

values = data['sex'].value_counts().tolist()
labels = data['sex'].value_counts().index.tolist()

print(values)
print(labels)

total = sum(values)

plt.pie(values, labels=labels, autopct=lambda p: '{:.0f}'.format(p * total / 100))
```

```python
data = pd.read_csv('tips.csv')

total_bill = data['total_bill'] 
tip = data['tip'] 

bins = [0,10,20,30,40]

plt.hist(tip, label="Tip", bins=bins, edgecolor="black")
plt.hist(total_bill, label="Bill", bins=bins, edgecolor="black")

plt.legend()
```

```python
data = pd.read_csv('tips.csv')

grouped = data.groupby('smoker')

yes = grouped['total_bill'].get_group("Yes")
no = grouped['total_bill'].get_group("No")

plt.hist(yes, label="Smoker", bins=5, edgecolor="black")
plt.hist(no, label="Non-smoker", bins=5, edgecolor="black")

plt.legend()

data.sort_values(by=['total_bill', 'smoker'], ascending=[True, False], inplace=True)
print(data.tail(10))
```
