# Susipažinkime su biblioteka. 
### plt.plot() 

plot naudojamas brėžti taškams/linijoms grafike

Šitame pavyzdyje tiesiog nubrėžiame linija. Pagal numatytuosius nustatymus funkcija plot() brėžia liniją nuo taško iki taško.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = (20,25)
y = (3,6)

plt.plot(x,y, markersize=12, marker='o')
```

Galima nubrėžti ir tiesiog du taškus. Taip pat šitame pavyzdyje paryškiname taškus su color

```python
x = [20,25]
y = [3,6]

plt.plot(x,y, "o", color="red")
```

```python
x = [20,25]
y = [3,6]

plt.title("Hello matplotlib")
plt.plot(x,y, label="test")

plt.grid() # uždedame langelius
plt.legend() # uždedame legendą
```

Galima liniją brėžti tik su y reikšmėmis, o x'ams bus priskirtos įprastinės reiškmės

```python
y = np.array([1,10])

plt.title("Hello matplotlib")
plt.plot(y, label="test")

plt.grid() # uždedame langelius
plt.legend() # uždedame legendą
```

Galima brėžti kelias linijas ar net kelis skirtingus vizualizacijos tipus (pažiurėsime vėliau)

```python
x = [20,25]
y = [3,6]

x2 = [40,45]
y2 = [13,16]

plt.title("Hello matplotlib")
plt.plot(x,y, label="test")
plt.plot(x2,y2, label="test2")

plt.grid() 
plt.legend() 
```

# Stulpelinė diagrama (bar chart) 
### plt.bar() 

```python
plt.style.use('seaborn-whitegrid')

x = 20
y = 3

plt.title("Hello matplotlib")
plt.bar(x,y, label="test", color="b")

plt.grid(True)
plt.legend(loc='upper left')
```

# Galime naudoti ir bar ir plot kartu

```python
plt.style.use('seaborn-whitegrid')

x = np.array([20,25])
y = np.array([3,6])

x2 = np.array([40,45])
y2 = np.array([13,16])

plt.title("Hello matplotlib")

plt.bar(x,y, label="test", color="b")
plt.bar(x2,y2, label="test2", color="r")
plt.plot(x,y, label="test", color="b", linestyle="--", marker=".")
plt.plot(x2,y2, label="test2", color="r", marker=".")

plt.grid(True)
plt.legend(loc='upper left')
```

```python
plt.style.use('seaborn-whitegrid')

x = np.array([20,25])
y = np.array([3,6])

x2 = np.array([40,45])
y2 = np.array([13,16])

plt.title("Hello matplotlib")

plt.bar(x2,y2, label="test2", color="r")
plt.plot(x2,y2, label="test2", color="r", marker=".")
plt.fill_between(x2,y2, color="red")

plt.bar(x,y, label="test", color="b")
plt.plot(x,y, label="test", color="b", linestyle="--", marker=".")
plt.fill_between(x,y, color="blue")

plt.grid(True)
plt.legend(loc='upper left')
```

# Pie Charts (Skritulinė diagrama)

```python
data = pd.read_csv('tips.csv')
total_bill = data['total_bill']
tip = data['tip']
values = [sum(total_bill), sum(tip)]
labels = ["Total bill", "Tips"]
explode = [0,0.1]

plt.pie(values, startangle=90, labels=labels, explode=explode, autopct='%1.1f%%')
```

## Pavaizduokime visus galimus porcijos dydžius

```python
data = pd.read_csv('tips.csv')

values = data['size'].value_counts().tolist()
labels = data['size'].value_counts().index.tolist()

print(values)
print(labels)

total = sum(values)

plt.pie(values, labels=labels, autopct=lambda p: '{:.0f}'.format(p * total / 100))
```

# Histogramos

Panaši į stulpelinę diagramą, tačiau galima grupuoti reikšmes į atskirus "paketus" (bins). Jų pagalba matysime, koks yra kokio nors vieno rodiklio pasiskirstymas.

```python
data = pd.read_csv('tips.csv')
total_bill = data['sex'] 

plt.hist(total_bill, label="Sex")
plt.legend()
```

```python
data = pd.read_csv('tips.csv')
total_bill = data['total_bill'] 

plt.hist(total_bill, label="Bill", bins=5, edgecolor="black")
plt.legend()
```

```python
data = pd.read_csv('tips.csv')

total_bill = data['total_bill'] 
tip = data['tip'] 

plt.hist(total_bill, label="Bill", bins=5, edgecolor="black")
plt.hist(tip, label="Tip", bins=5, edgecolor="black")
plt.legend()
```

```python
data = pd.read_csv('tips.csv')

total_bill = data['total_bill'] 
tip = data['tip'] 

bins = [0,10,20,30,40]

plt.hist(tip, label="Tip", bins=bins, edgecolor="black")
plt.hist(total_bill, label="Bill", bins=bins, edgecolor="black")

plt.legend()
```

## Sugrupuokime Bill pasiskirstymą tarp rūkančių ir nerūkančių ir nupieškime histogramą

```python
data = pd.read_csv('tips.csv')

grouped = data.groupby('smoker')

yes = grouped['total_bill'].get_group("Yes")
no = grouped['total_bill'].get_group("No")

plt.hist(no, label="Non-smoker", bins=5, edgecolor="black")
plt.hist(yes, label="Smoker", bins=5, edgecolor="black")

plt.legend()

data.sort_values(by=['total_bill', 'smoker'], ascending=[True, False], inplace=True)
print(data.tail(10))
```
